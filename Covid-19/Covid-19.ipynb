```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('covid_19_india _dataset.csv')
x = dataset.iloc[:,0].values
y = dataset.iloc[:,-1].values
```

```python
plt.scatter(x, y, color = 'red')
plt.title('Covid-19 cases')
plt.xlabel('Day')
plt.ylabel('Total Cases')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
x = x.reshape(-1,1)
regressor = PolynomialFeatures(degree = 4)
x_train_poly = regressor.fit_transform(x)
lin_regressor = LinearRegression()
lin_regressor.fit(x_train_poly, y)
y_pred_poly = lin_regressor.predict(regressor.fit_transform(x))
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_regressor.predict(regressor.fit_transform(x)), color = 'blue')
plt.title('Covid-19 cases')
plt.xlabel('Day')
plt.ylabel('Total Cases')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_pred_poly)
```
